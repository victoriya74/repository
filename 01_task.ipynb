1. Загрузить, посмотреть, определить количество строк и объединить 3 датасета: marketing_campaign.csv, users.csv и subscribers.csv.

```python
#Считаем файлы
import pandas as pd

DATASET_PATH1 = './marketing_campaign.csv'
DATASET_PATH2 = './subscribers.csv'
DATASET_PATH3 = './users.csv'
df_marketing = pd.read_csv(DATASET_PATH1, sep=',')
df_subscribers = pd.read_csv(DATASET_PATH2, sep=',')
df_users = pd.read_csv(DATASET_PATH3, sep=',')
```

```python
# Объединим три таблицы по полю user_id
```

```python
df_merge1 = pd.merge( df_marketing, df_subscribers , on ='user_id' , how ='inner' )
df_merge2 = pd.merge(df_merge1, df_users, on ='user_id', how='inner')
df_merge2.head()
```

3. Определить эффективность маркетинговых каналов по привлечению платящих игроков

```python
#отфильтруем по условию, что клиент совершившил покупку
#и посмотрим в группировке по рекламному каналу кол-во клиентов
df =  df_merge2[df_merge2['converted'] == 1]
df2 = df.groupby('marketing_channel').count()['user_id']

df2.sort_values( ascending=False)
#по абсолютному значению клиентов больше у маркетинговый канал  "House Ads"
```

```python
#посчитаем относительные значения: кол-во клиентов в канале, совершивших покупку к общему кол-ву в этом канале
```

```python
df.groupby('marketing_channel').count()['user_id']/df_merge2.groupby('marketing_channel').count()['user_id']*100
#по относительному значению эффективнее канал Email
```
